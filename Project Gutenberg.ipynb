# Project Gutenberg 

We can get a few texts from project gutenberg like this, but we are going to use the lighter version gutenbergpy going forward. 

```python
import re
import itertools
import collections
import nltk
import urllib3
import numpy as np
from pylab import rcParams
from nltk.corpus import stopwords

nltk.download('punkt_tab')
```

```python
import gutenbergpy.textget as tg

def gettext(id):

    raw_book   = tg.get_text_by_id(id)
    clean_book = tg.strip_headers(raw_book)

    return clean_book, raw_book
```
